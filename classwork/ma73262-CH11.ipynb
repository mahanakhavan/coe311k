## Cubic Spline vs Polynomial Fitting

```python
import numpy as np

import matplotlib.pyplot as plt

from scipy.interpolate import CubicSpline

 


x = np.array([0, 1, 2, 3, 4, 5])

y = np.array([0, 1, 0, 1, 0, 1])



cs = CubicSpline(x, y, bc_type='natural')

x_interp = np.linspace(0, 5, 100)

y_interp = cs(x_interp)

degree = 5
coeffs = np.polyfit(x, y, degree)
poly = np.poly1d(coeffs)
y_polyfit = poly(x_interp)
```

## Plot

```python
plt.plot(x, y, 'o', label='Data Points')
plt.plot(x_interp, y_interp, label='CubicSpline')
plt.plot(x_interp, y_polyfit, label='Polyfit')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Comparison of Polyfit and CubicSpline')
plt.legend()
plt.grid(True)
plt.show()
```

We can see that the CubicSpline interpolation graph is a consistant smooth curve, with oscillation that resemble any sort of sine or cosine graph, showing a uniform structure. The Polyfit interpolation graph, which uses a single high degree polynomial, has wider, less uniform oscillations. This can especially be seen near the boundaries. Overall, we can see that the CubicSpline is more stable for the interpolation of multiple data points.
