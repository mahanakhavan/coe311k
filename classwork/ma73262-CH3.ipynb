```python
import math
import random
import numpy as np

def montecarlo_pi(number_of_shots):
    number_of_hits = 0
    for r in range(0, number_of_shots):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if(x*x + y*y < 1):
            number_of_hits = number_of_hits + 1
            
    return 4*number_of_hits/number_of_shots
```

```python
absoluteErrors = []
relativeErrors = []

for i in range (0,100):
    pi = montecarlo_pi(1000)
    truePi = np.pi
    absoluteError = np.absolute(truePi - pi)
    relativeError = np.absolute((truePi - pi)/truePi)
    
    absoluteErrors.append(absoluteError)
    relativeErrors.append(relativeError)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(absoluteErrors)
plt.xlabel("Run Number")
plt.ylabel("Absolute Error")
plt.title("Absolute Error of Monte Carlo Pi Estimate")
plt.show()

plt.figure()
plt.plot(relativeErrors)
plt.xlabel("Run Number")
plt.ylabel("Relative Error")
plt.title("Relative Error of Monte Carlo Pi Estimate")
plt.show()
```
