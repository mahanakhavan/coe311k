# Leibniz Series for Pi 

```python
import numpy as np
```

#### What's happening here is that the error values are slowly but surely decreasing in size as the number of samples increases. Additionally, the values are actually alternating between positive and negative, due to the series being an alternating series. You can't really see this though, since we're only displaying every 100 "n," which are all even numbers. However, we acn still see the error getting smaller and smaller.


```python
true_pi = np.pi

pi_over_4_sum = 0
for n in range(1, 5001):
    term = ((-1)**(n-1)) / (2*n - 1)   
    pi_over_4_sum += term
    pi_est = 4 * pi_over_4_sum
    
    if n % 100 == 0:
        rel_error = abs((true_pi - pi_est)/true_pi)
        print("n =", n, "pi_est =", pi_est, "relative error =", rel_error)
```
