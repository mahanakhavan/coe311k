```python
import numpy as np

EPS = 1e-12
```

## Gaussian Elimination with pivoting

```python
def elim_pivot(A, b):
    A = np.array(A, float)
    b = np.array(b, float)
    n = len(b)

    for k in range(n - 1):
        r = k + np.argmax(np.abs(A[k:, k]))
        if abs(A[r, k]) < EPS:
            raise ValueError("Singular matrix")

        if r != k:
            A[[k, r]] = A[[r, k]]
            b[[k, r]] = b[[r, k]]

        for i in range(k + 1, n):
            m = A[i, k] / A[k, k]
            A[i, k] = 0.0
            A[i, k + 1:] = A[i, k + 1:] - m * A[k, k + 1:]
            b[i] = b[i] - m * b[k]

    if abs(A[n - 1, n - 1]) < EPS:
        raise ValueError("Singular matrix")

    return A, b
```

## Back Substitution 

```python
def back_sub(U, b):
    n = len(b)
    x = np.zeros(n)

    for i in range(n - 1, -1, -1):
        s = np.dot(U[i, i + 1:], x[i + 1:])
        x[i] = (b[i] - s) / U[i, i]

    return x
```

## Solve Gaussian

```python
def gauss_solve(A, b):
    U, bb = elim_pivot(A, b)
    return back_sub(U, bb)
```

## LU Pivoting

```python
def lu_inplace(A):
    A = np.array(A, float)
    n = A.shape[0]
    p = np.arange(n)

    for k in range(n - 1):
        r = k + np.argmax(np.abs(A[k:, k]))
        if abs(A[r, k]) < EPS:
            raise ValueError("Singular matrix")

        if r != k:
            A[[k, r]] = A[[r, k]]
            p[[k, r]] = p[[r, k]]

        for i in range(k + 1, n):
            A[i, k] = A[i, k] / A[k, k]
            A[i, k + 1:] = A[i, k + 1:] - A[i, k] * A[k, k + 1:]

    return A, p
```

## LU Solve

```python
def lu_solve(A_lu, p, b):
    b = np.array(b, float)
    n = len(b)

    bp = b[p]

    y = np.zeros(n)
    for i in range(n):
        y[i] = bp[i] - np.dot(A_lu[i, :i], y[:i])

    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        s = np.dot(A_lu[i, i + 1:], x[i + 1:])
        x[i] = (y[i] - s) / A_lu[i, i]

    return x
```

## Example

```python
A = [[2, 1, -1],
     [-3, -1, 2],
     [-2, 1, 2]]
b = [8, -11, -3]

print("Gaussian:", gauss_solve(A, b))

A_lu, p = lu_inplace(A)
print("LU:", lu_solve(A_lu, p, b))
```
