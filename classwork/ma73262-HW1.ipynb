# Homework 1: Plot Monte Carlo Pi 

### Monte Carlo Pi Function

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt

def montecarlo_pi(number_of_shots):
    number_of_hits = 0
    for r in range(0, number_of_shots):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if(x*x + y*y < 1):
            number_of_hits = number_of_hits + 1
            
    return 4*number_of_hits/number_of_shots
```

### Main: Iterate Samples

```python
sample_sizes = range(10, 1001, 20)
absolute_errors = []

true_pi = np.pi

for n in sample_sizes:
    pi_estimate = montecarlo_pi(n)
    error = abs(pi_estimate - true_pi)
    absolute_errors.append(error)         #to make sure it stores all the values and puts them all together
```

### Plot

```python
plt.figure()
plt.plot(sample_sizes, absolute_errors, marker='o') #just to show the jumps by 20
plt.xlabel("Number of Samples")
plt.ylabel("Absolute Error")
plt.title("Monte Carlo Pi Approximation Error vs. Number of Samples")
plt.grid(True)
plt.show()
```
