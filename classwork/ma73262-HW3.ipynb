# Taylor Series, Truncation Error

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

### Taylor Series For Sin Function

```python
def sin_taylor(x, n_terms):
    s = 0.0
    for k in range(n_terms):
        power = 2 * k + 1
        term = ((-1) ** k) * (x ** power) / math.factorial(power)
        s += term
    return s
```

### Defining Variables

```python
x = np.arange(0.0, 10.1, 0.1)

y_exact = np.sin(x)

term_counts = [1, 2, 3, 4]
errors = []
```

### Running Loop

```python
for n in term_counts:
    y_approx = []
    for val in x:
        y_approx.append(sin_taylor(val, n))
    y_approx = np.array(y_approx)
    
    abs_error = np.abs(y_approx - y_exact)
    errors.append(abs_error)
```

### Plot

```python
plt.figure()
plt.plot(x, errors[0])
plt.plot(x, errors[1])
plt.plot(x, errors[2])
plt.plot(x, errors[3])

plt.xlabel("x")
plt.ylabel("absolute error")
plt.title("Taylor Series Truncation Error for sin(x) using 1 to 4 terms")
plt.legend(["1 term", "2 terms", "3 terms", "4 terms"])
plt.grid(True)
plt.show()
```

### Discuss:
We can see that, the truncation error appears actually larger for higher term Taylor-Series approximations despite higher terms supposedly signifying greater accuracy. While it is true, the more terms the more accuracy, what's happening here is that the Taylor series is centered at x = 0, and we're starting to use very large x values, that are blowing up exponentially due to the higher powers present in the higher term Taylor Series. Sin is bounded between -1 and 1, and while more terms will certainly increase accuracy for terms closer to x = 0, the error will start to blow up until plenty of terms have been added.
